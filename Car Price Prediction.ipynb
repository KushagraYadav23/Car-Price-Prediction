```python
# 2. DATA PROCESSING

```

```python
print(df.columns)
```

```python
newdf=df.drop(columns=['customer name','customer e-mail', 'gender', 'country'])    # DROPPING UNWANTED COLUMNS and saving data in new df
newdf.head(10)
```

```python
newdf.corr()       # Tells us how strongly the columns are linearly related to each other
```

```python
newdf.drop(columns=['credit card debt'], inplace=True)    # droping 'credit card dept' as it is not affecting data to much
newdf.head(10)
```

```python
newdf.shape
```

```python
# define features(X) and target(Y)
```

```python
x=newdf.drop(columns=['car purchase amount'])
y=newdf[['car purchase amount']]

print("The size of the INPUT training data: ",x.shape)
print("The size of the OUTPUT training data: ",y.shape)
```

```python

x.head()
```

```python
y.head()
```

```python
scaler= MinMaxScaler()
x=scaler.fit_transform(x)
y=scaler.fit_transform(y)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1 , random_state = 25)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# MACHINE LEARNING MODLE
```

```python
lr_model=LinearRegression()          # calling the linear regression model in variable and then fitting the training data in the model
lr_model.fit(x_train,y_train)
```

```python
rf_model=RandomForestRegressor()
rf_model.fit(x_train,y_train)
```

```python
gbr_model=GradientBoostingRegressor()
gbr_model.fit(x_train,y_train)
```

```python
# prediction on test data
```

```python
lr_pred=lr_model.predict(x_test)
rf_pred=rf_model.predict(x_test)
gbr_pred=gbr_model.predict(x_test)
```

```python
# EVALUATING: checkinng the accuracy of our model using different techniques like r2 score and mean squared error
```

```python
lr_score = metrics.r2_score(y_test,lr_pred)
rf_score = metrics.r2_score(y_test,rf_pred)
gbr_score = metrics.r2_score(y_test,gbr_pred)

print(lr_score,rf_score,gbr_score)
```

```python
model_data = pd.DataFrame({'Models':['LR','RF','GBR'],
             "R2_SCORE":[lr_score,rf_score,gbr_score]})
```

```python
model_data
```
